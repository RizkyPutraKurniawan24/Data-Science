### **Comparison Operator : < , > , == , <= , >=**

```python
# Comparison operator
x = 10
y = 24

# apakah x lebih kecil dari y?
a = x < y
print (f'a < b : {a}')

# apakah x lebih besar dari y?
b = x > y
print (f'a > b : {b}')

# apakah x sama dengan y?
c = x == y
print (f'x == y : {c}')

# apakah x lebih kecil sama dengan dari y?
d = x <= y
print (f'a <= b : {d}')

# apakah x lebih besar sama dengan dari y?
e = x >= y
print (f'a >= b : {e}')
```

## **2.2 SCOPE**

Apa yang perlu diketahui tentang scope:
*   global: bagian utama yang ditentukan dalam script
*   local: didefinisikan dalam suatu fungsi
*   built in scope: nama dalam scope module bawaan yang telah ditentukan sebelumnya seperti print, len

```python
x = 2
def f():
    x = 3
    return x
print(x)      # x = 2 global scope
print(f())    # x = 3 local scope
```

```python
# no local scope
x = 5
def f():
    y = 2*x        # there is no local scope x
    return y
print(f())         # it uses global scope x
# First local scope searched, then global scope searched, if two of them cannot be found lastly built in scope searched.
```

```python
# built in scope
import builtins
dir(builtins)
```

## **2.3 NESTED FUNCTION**

*   function didalam function
*   Ada LEGB rule yaitu pencarian local scope, enclosing function, global dan built in scopes,masing-masing.

```python
def square():
    """ return square of value """
    def add():
        """ add two local variable """
        x = 2
        y = 3
        z = x + y
        return z
    return add()**2
print(square())
```

## **2.4 DEFAULT and FLEXIBLE ARGUMENTS**


### **Default Arguments**

```python
def f(a, b = 1, c = 2):
    y = a + b + c
    return y
print(f(5))
# what if we want to change default arguments
print(f(5,4,3))
```

### **Flexible Arguments (args)**

```python
def f(*args):
    for i in args:
        print(i)
f(1)
print("")
f(1,2,3,4)
```

### **Flexible Arguments (kwargs)**

```python
def f(**kwargs):
    """ print key and value of dictionary"""
    for key, value in kwargs.items(): # If you do not understand this part turn for loop part and look at dictionary in for loop
        print(key, " ", value)
f(country = 'Spain', capital = 'Madrid', population = 123456)
```

### **LIST COMPREHENSION**

List Comprehension : collapse loop untuk membuat daftar menjadi satu baris

Contoh: num1 = [1,2,3] dan kita ingin menjadikannya num2 = [2,3,4]. Ini dapat dilakukan dengan perulangan for. Namun itu terlalu panjang. Kita dapat menjadikannya satu kode baris yaitu list comprehension.

```python
num1 = [1,2,3]
print(num1)
num2 = [i + 1 for i in num1 ]
print(num2)
```

[i + 1 for i in num1 ] : list of comprehension

i +1 : list comprehension syntax

for i in num1: for loop syntax

i: iterator

num1: iterable object

```python
# Conditionals on iterable
num1 = [5,10,15]
print(num1)
num2 = [i**2 if i == 10 else i-5 if i < 7 else i+5 for i in num1]
print(num2)
```

## **4.6 RESAMPLING PANDAS TIME SERIES**
*   Resampling : Statistical method pada interval waktu yang berbeda
>*   Membutuhkan string untuk menentukan frekuensi seperti "M" = bulan atau "A" = tahun
*   Downsampling : Kurangi datetime rows ke frekuensi yang lebih lambat seperti dari harian ke mingguan
*   Upsampling : Tingkatkan datetime rows ke frekuensi yang lebih cepat seperti dari harian menjadi setiap jam
*   Interpolate : Interpolate nilai menurut metode berbeda seperti 'linier', 'time', atau 'index'

```python
# We will use data2 that we create at previous part
data2.resample("A").mean()
```

```python
# Lets resample with month
data2.resample("M").mean()
# As you can see there are a lot of nan because data2 does not include all months
```

```python
# In real life (data is real. Not created from us like data2) we can solve this problem with interpolate
# We can interpolete from first value
data2.resample("M").first().interpolate("linear")
```

```python
# Or we can interpolate with mean()
data2.resample("M").mean().interpolate("linear")
```
